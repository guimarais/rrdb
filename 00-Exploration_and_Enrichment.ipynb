```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
df = pd.read_excel("./data/ReactorList.xlsx", skiprows=1)
```

```python
df.head()
```

```python
df['First Criticality Date'] = pd.to_datetime(df['First Criticality Date'])
```

```python
df.head()
```

```python
cc = pd.read_csv("./data/Countries-Continents.csv")
```

```python
cc.head()
```

```python
dfc = df.merge(cc, on=['Country'], how='left')
```

```python
alt.Chart(dfc).mark_point().encode(
    x='First Criticality DateSyrian Arab Republic',
    y='Status:O',
    color='Continent',
    tooltip=['Facility Name', 'City', 'Country', 'Type']
).properties(
    width=800,
    height=200
).interactive()
```
